```python
import pandas as pd
import numpy as np
```

```python
x = pd.read_csv('ds_salaries.csv')
cleaned_data = x.copy()
```

```python
cleaned_data
```

```python
cleaned_data.rename(columns={'Unnamed: 0':'Index'}, inplace=True)
```

```python
cleaned_data
```

```python
def remote_ratio_changer(value):
    if value == 0:
        return 'Office'
    elif value == 50:
        return 'Partially Remote'
    elif value == 100:
        return 'Fully Remote'
    else:
        return 'Unknown'
```

```python
cleaned_data['remote_ratio'] = cleaned_data['remote_ratio'].apply(remote_ratio_changer)
```

```python
cleaned_data
```

```python
cleaned_data.rename(columns={'remote_ratio':'work_location'}, inplace=True)
```
